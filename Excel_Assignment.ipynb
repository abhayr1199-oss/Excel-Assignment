Question 1: What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly
explain with examples.


Answer:

Paste inserts everything copied — values, formulas, formatting, and comments.
Example:
If cell A1 contains the formula =B1+C1, using Paste will paste the formula as it is into the destination cell.

Paste Special allows you to paste selected attributes — only values, only formulas, only formatting, etc.
Example:
If A1 contains =B1+C1, but you only want the calculated number (not the formula):
Use Paste Special → Values.

Other Paste Special options:

Paste Values

Paste Formats

Paste Formulas

Paste Comments

Paste Transpose

Question 4: What are logical functions in Excel? Provide examples of at least two
logical functions and their applications.


Answer:

Logical functions return TRUE or FALSE based on a condition.

Examples:

| Function | Example                    | Result                                   |
| -------- | -------------------------- | ---------------------------------------- |
| IF       | `=IF(A1>50,"Pass","Fail")` | Returns Pass if value>50                 |
| AND      | `=AND(A1>0,A1<100)`        | Returns TRUE if both conditions are true |
| OR       | `=OR(A1=10,A1=15)`         | TRUE if either condition matches         |

Applications: Data validation, decision making, filtering results, conditional formatting.

Question 5: Discuss the purpose of 'XLOOKUP' and how it differs from the traditional
'VLOOKUP' function.


Answer:

XLOOKUP Purpose

Searches for a lookup value in a range and returns related value from another range.
| Feature                | VLOOKUP                          | XLOOKUP                          |
| ---------------------- | -------------------------------- | -------------------------------- |
| Direction              | Only searches left → right       | searches both directions         |
| Column number          | Must specify col number manually | selects return column directly   |
| Exact match default    | No (approx default)              | Yes                              |
| Handles missing values | No                               | Yes (can specify default result) |

Example:

=XLOOKUP(A2, A10:A20, B10:B20, "Not Found")

Question 10: You're working with a dataset that contains missing values. As a Data
Scientist, explain how you'd detect and handle missing data using Excel.

Mention tools like:

● Go To Special

● ISBLANK

● COUNTBLANK

(Include a screenshot showing how blanks are identified or processed.)

Answer:

Missing values cause incorrect analysis. In Excel, we detect and handle them using:

Method 1: Go To Special

1. Select the full sheet or range

2. Press Ctrl + G → Go To

3. Click Special

4. Select Blanks

Excel highlights all blank cells.

Method 2: ISBLANK function

Used to test if a cell is blank.

Example:
=ISBLANK(A2) → returns TRUE if A2 is empty

Method 3: COUNTBLANK function

Counts number of blank cells in a range.

Example:
=COUNTBLANK(A1:A20)


Handling missing data options

Fill blanks with 0 or NA

Delete blank rows

Replace blanks using Fill Down

Use conditional formatting to highlight blanks
